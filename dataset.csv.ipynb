```python
import pandas as pd
import random

# Base tweets
positive = [
    "MS Dhoni is the greatest captain ever!", "Thala for a reason!", "Mahi's helicopter shot is iconic.",
    "Best finisher in the history of cricket.", "Dhoni's wicketkeeping skills are unmatched.",
    "I have massive respect for MS Dhoni.", "CSK is nothing without Thala.", "Dhoni won us the World Cup!",
    "No one can replace MS Dhoni in Indian Cricket.", "Mahi is my biggest inspiration."
]
negative = [
    "Dhoni should have retired much earlier.", "His strike rate was too low in the last few years.",
    "I feel MS Dhoni is a bit overrated.", "He played too slow in the 2019 semi-final.",
    "Not a fan of his test captaincy.", "Dhoni took the spot of younger players.",
    "Too much unnecessary hype around Dhoni.", "His defensive strategy cost us matches sometimes.",
    "I don't like how CSK relies only on Dhoni.", "He was a good player but past his prime now."
]
neutral = [
    "MS Dhoni played for the Indian Cricket Team.", "Dhoni's jersey number is 7.",
    "Mahi was born in Ranchi.", "MS Dhoni is the captain of Chennai Super Kings.",
    "I saw a news article about Dhoni today.", "Dhoni announced his retirement on August 15.",
    "There is a biopic movie made on MS Dhoni.", "Dhoni hit a six in yesterday's match.",
    "MSD has played many ICC tournaments.", "Dhoni was at the event in Mumbai."
]

tweets = []
labels = []

# Generate exactly 100 tweets (40 Positive, 30 Negative, 30 Neutral)
for i in range(40):
    tweets.append(random.choice(positive) + f" ({i+1})")
    labels.append("Positive")
for i in range(30):
    tweets.append(random.choice(negative) + f" ({i+1})")
    labels.append("Negative")
for i in range(30):
    tweets.append(random.choice(neutral) + f" ({i+1})")
    labels.append("Neutral")

# Create DataFrame, shuffle it, and save to CSV
df = pd.DataFrame({"Tweet": tweets, "Sentiment": labels})
df = df.sample(frac=1).reset_index(drop=True)
df.to_csv("dataset.csv", index=False)

print("Success! 100 Tweets generated. Check the folder icon on the left to download dataset.csv.")
```
